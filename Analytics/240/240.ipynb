```python
f = open('./coins.in', 'r')
l = [line.strip() for line in f]
l[9995]
```

```python
import scipy.stats as stats

def clopper_pearson_interval(k, n, alpha=0.05):
    lower = stats.beta.ppf(alpha / 2, k, n - k + 1)
    upper = stats.beta.ppf(1 - alpha / 2, k + 1, n - k)
    return 0 if np.isnan(lower) else lower, 1 if np.isnan(upper) else upper
```

```python
intervals = []
for line in l[1:]:
    flips, heads = map(int, line.split())
    interval = clopper_pearson_interval(k=heads, n=flips, alpha=0.99)
    intervals.append((interval[1] + interval[0]) / 2)
```

```python
enumerated = [[intervals[i], i] for i in range(len(intervals))]
en1 = sorted(enumerated, key= lambda item : item[0] , reverse=True)
ans = []
for i in np.array(en1)[:, 1]:
    ans.append(int(i))
```
