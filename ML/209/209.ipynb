```python
import numpy as np
from catboost import CatBoostClassifier, Pool
```

```python
import random
def RandomPermutation():
    perm = list(range(8))
    random.shuffle(perm)
    return perm
```

```python
def StupidPermutation():
    partialSums = [0,1,8,35,111,285,
        628,1230,2191,3606,5546,8039,11056,14506,18242,
        22078,25814,29264,32281,34774,36714,38129,39090,
        39692,40035,40209,40285,40312,40319,40320]
    r = random.randint(0, partialSums[-1])
    numInv = 0
    while partialSums[numInv] < r:
        numInv += 1
    perm = list(range(8))
    for step in range(numInv):
        t1 = random.randint(0, 7)
        t2 = random.randint(0, 7)
        perm[t1], perm[t2] = perm[t2], perm[t1]
    return perm
```

```python
from tqdm import tqdm
```

```python
# each object has 64 features - flatten 8x8 matrix of counting how many times each digin from 0 to 7 was on each place on 1000 permutations
def generate_sample(size):
  x = []
  y = []
  for i in tqdm(range(size)):
    label = random.randint(0, 1)
    perm = np.zeros((8, 8))
    if label == 1:
      for i in range(1000):
        sample = RandomPermutation()
        for i in range(8):
          perm[i][sample[i]] += 1
    else:
      for i in range(1000):
        sample = StupidPermutation()
        for i in range(8):
          perm[i][sample[i]] += 1
    x.append(perm.flatten())
    y.append(label)
  return np.array(x), np.array(y)
```

```python
x_train, y_train = generate_sample(1000)
```

```python
x_test, y_test = generate_sample(200)
```
